```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
apartamente = [(37, 65), (55, 80), (80, 120), (60, 95), (92, 205), (41, 75), (170, 400), (145, 367), (82, 205), (72, 198)]

apartamente = np.array(apartamente)
apartamente = apartamente[apartamente[:,0].argsort()]
```

```python
suprafete, preturi = apartamente[:, 0], apartamente[:, 1]
```

```python
suprafete
```

```python
plt.scatter(suprafete, preturi)
plt.xlabel("Suprafata")
plt.ylabel("Preturi");
```

```python
plt.scatter(suprafete, preturi)
plt.plot(suprafete, preturi)
plt.xlabel("Suprafata")
plt.ylabel("Preturi");
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
suprafete.reshape(-1, 1)
```

```python
model.fit(suprafete.reshape(-1, 1), preturi)
```

```python
model.coef_, model.intercept_
```
