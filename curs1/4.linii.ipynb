```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    a = 0
    b = 0
    return a * x + b

suprafete = np.arange(40, 51)
suprafete
```

```python
preturi = []

for s in suprafete:
    preturi.append(f(s))

print(preturi)
```
